```python
import pickle

_ARCH = "binResNet"
_SUFFIX = ""
today = "20-06-23"

filename_los = "./training_lists/" + _ARCH + "los" + _SUFFIX + today + '.data'
filename_acc = "./training_lists/" + _ARCH + "acc" + _SUFFIX + today + '.data'
filename_vallos = "./training_lists/" + _ARCH + "vallos" + _SUFFIX + today + '.data'
filename_valacc = "./training_lists/" + _ARCH + "valacc" + _SUFFIX + today + '.data'

with open(filename_los, 'rb') as filehandle:
    loss_list = pickle.load(filehandle)
with open(filename_acc, 'rb') as filehandle:
    accuracy_list = pickle.load(filehandle)
with open(filename_vallos, 'rb') as filehandle:
    val_loss_list = pickle.load(filehandle)
with open(filename_valacc, 'rb') as filehandle:
    val_acc_list = pickle.load(filehandle)
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec


fig = plt.figure(figsize=(8, 6), dpi=144, constrained_layout=True)
gs = gridspec.GridSpec(2, 1, figure=fig)
xlim = len(loss_list) + 1
x = np.arange(1, xlim)

ax1 = fig.add_subplot(gs[0,0])
ax1.set_xlim(-1, xlim)
ax1.set_ylim(0, 1.02)
ax1.set_xlabel("Epochs")
ax1.set_ylabel("Accuracy")
ax1.plot(x, accuracy_list, linestyle='-', label='training')
ax1.plot(x, val_acc_list, linestyle='-', label='validation')
plt.yticks(np.arange(0, 1.02, 0.1))
plt.grid(True)
plt.title('Accuracy')
ax1.legend()

ax2 = fig.add_subplot(gs[1,0])
ax2.set_xlim(-1, xlim)
ax2.set_ylim(0, 2.04)
ax2.set_xlabel("Epochs")
ax2.set_ylabel("Loss")
ax2.plot(x, loss_list, linestyle='-', label='training')
ax2.plot(x, val_loss_list, linestyle='-', label='validation')
plt.yticks(np.arange(0, 2.04, 0.2))
plt.grid(True)
plt.title('Loss')
ax2.legend()

plt.draw()
plt.show()
```
